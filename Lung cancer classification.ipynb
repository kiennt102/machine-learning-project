```python
data_for_model = df[[
    'Air Pollution',
    'Alcohol use',
    'Dust Allergy',
    'OccuPational Hazards',
    'Genetic Risk',
    'chronic Lung Disease',
    'Balanced Diet',
    'Obesity',
    'Smoking',
    'Passive Smoker',
    'Chest Pain',
    'Coughing of Blood',
    'Fatigue',
    'Level'
]]
```

```python
train_val_set = data_for_model[:900]
test_set = data_for_model[900:]
```

```python
# Take X and y
```

```python
X = train_val_set.drop(['Level'], axis=1)
print("X for train_val_set : ", len(X))
X.head(2)
```

```python
y = train_val_set['Level']
print("y for train_val_set : ", len(y))
y.head(2)
```

```python
# TAKE FEATURE FOR TRAINING X_train, y_train, X_test, y_test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=43)
```

```python
y_train.value_counts()
```

```python
y_val.value_counts()
```

```python
# USING LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression( penalty=None,C=0.01, solver="saga")
logistic_model.fit(X_train, y_train)
```

```python
# Take prediction and mesure ACC, RECAL, PREC
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, confusion_matrix
```

```python
y_pred_log = logistic_model.predict(X_val)

print(f"Accuracy : {accuracy_score(y_true=y_val, y_pred=y_pred_log):0.4f}")
print("\n")
print(classification_report(y_true=y_val, y_pred=y_pred_log))
```

```python
cf_matrx_train = confusion_matrix(y_train, logistic_model.predict(X_train))

plt.title("LogisticRegression confusion matrix on train set")
sns.heatmap(cf_matrx_train, annot=True,cmap="Blues",fmt="d" );
```

```python
cf_matrx_val = confusion_matrix(y_val, y_pred_log)

plt.title("LogisticRegression confusion matrix on valid set")
sns.heatmap(cf_matrx_val, annot=True,cmap="Blues",fmt="d" );
```

```python
from sklearn.metrics import auc, roc_curve
from sklearn.preprocessing import label_binarize
```

```python
# Binarize label (Low, Medium, High)
y_test_binarized = label_binarize(y_test, classes=[0,1,2])
n_classes = y_test_binarized.shape[1]


probs = logistic_model.predict_proba(X_test)

# Calculate ROC and AUC for each classes
fpr, tpr, roc_auc = {}, {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Vẽ ROC curve cho từng lớp
plt.figure()
colors = ['blue', 'red', 'green']
classes = ['Low', 'Medium', 'High']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f"ROC curve of class {classes[i]} (AUC = {roc_auc[i]:.2f})")


plt.plot([0, 1], [0, 1], '--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Multiclass Classification')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


# Khởi tạo Logistic Regression
logistic_model = LogisticRegression(max_iter=1000)

# Định nghĩa grid các tham số để tìm kiếm
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# Khởi tạo GridSearchCV
grid_search = GridSearchCV(estimator=logistic_model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Kết quả tốt nhất
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Dự đoán trên tập test
y_pred = grid_search.best_estimator_.predict(X_val)

# Đánh giá mô hình
print("\nClassification Report:\n")
print(classification_report(y_val, y_pred))
```
